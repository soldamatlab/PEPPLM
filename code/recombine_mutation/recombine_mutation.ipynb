Save / Load results

```python
import pickle
file_name = 'results.pkl'
```

```python
with open(file_name, 'wb') as f:
    pickle.dump([best_variants, fitness_progressions], f)
```

```python
with open(file_name, 'rb') as f:
    best_variants, fitness_progressions = pickle.load(f)
```

Plot learning curve

```python
import matplotlib.pyplot as plt
```

```python
run_idx = 2

n_evaluations = [i for i in range(1, len(fitness_progressions[run_idx])+1)]
plt.plot(n_evaluations, fitness_progressions[run_idx])
plt.xlabel('Number of Evaluations')
plt.ylabel('Highest Fitness Reached (relative to WT variant)')
finish = 'Final variant: ' + best_variants[run_idx] + ' ({:.2f})'.format(fitness_progressions[run_idx][-1])
plt.title('Recombine mutation - GB1\n' + finish)
plt.show()
```
